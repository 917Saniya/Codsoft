```python
data_titanic = data_titanic.drop(columns=['Name', 'Ticket'], axis=1)
data_titanic.head()
```

```python
data_titanic["Sex"].value_counts()
```

```python
data_titanic['Embarked'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
cols = ['Sex', 'Embarked']
le = LabelEncoder()

for col in cols:
    data_titanic[col] = le.fit_transform(data_titanic[col])
data_titanic.head()
```

```python
X = data_titanic.drop(columns = ['PassengerId','Survived'],axis=1)
Y = data_titanic['Survived']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()

data_titanic.info()

```

```python
# Handle missing values before casting
data_titanic['Age'] = data_titanic['Age'].fillna(0)  # Replace NaN with 0, for instance
data_titanic = data_titanic.astype({'Age':'int','Fare':'int'})
print(data_titanic.dtypes)
```

```python
# Assuming 'Cabin' is the problematic column, replace non-numeric values with NaN
X_train['Cabin'] = pd.to_numeric(X_train['Cabin'], errors='coerce')

# Impute missing values (NaN) with a suitable strategy, such as the mean
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train)

# Now try fitting the model again
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
print(X_train_prediction)

```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy_score_of_training_data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)

print(X_test_prediction)

```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy_score_of_test data : ', test_data_accuracy)
```
