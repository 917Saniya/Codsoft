**SALES PREDICTION**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_Sales = pd.read_csv("/content/advertising(1).xls")
data_Sales.head()
```

```python
data_Sales.shape
```

```python
data_Sales.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Melt the DataFrame
data_melted = pd.melt(data_Sales, id_vars=['Sales'], value_vars=['TV', 'Radio', 'Newspaper'], var_name='Advertising Channel', value_name='Spending')

# Group the data and calculate total spending for each channel
channel_spending = data_melted.groupby('Advertising Channel')['Spending'].sum()

# Create a pie chart
plt.pie(channel_spending, labels=channel_spending.index, autopct='%1.1f%%')
plt.title('Total Spending by Advertising Channel')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_Sales[['TV']], data_Sales[['Sales']], test_size = 0.2,random_state=0)
print(X_train)

```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
res= model.predict(X_test)
print(res)
```

```python
print("Accuracy Score: ", model.score(X_test,y_test)*100)
```
